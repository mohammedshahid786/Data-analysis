```python
# 1. Import required libraries
import pandas as pd

# 2. Load the dataset
file_path = "/content/piza dataset.csv"
df = pd.read_csv(file_path)

# 3. Preview the dataset
print("Initial Dataset Preview:")
print(df.head())

# 4. Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# 5. Drop duplicate rows (if any)
df.drop_duplicates(inplace=True)

# 6. Drop rows with any missing values
df.dropna(inplace=True)

# 7. Convert 'Order Date' to datetime if present
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
    df.dropna(subset=['Order Date'], inplace=True)  # Drop rows where date couldn't be parsed

# 8. Calculate Total if Quantity and Price columns are present
if 'Quantity' in df.columns and 'Price' in df.columns:
    df['Total'] = df['Quantity'] * df['Price']

# 9. Sort the data (by Order Date if exists, else by Total or another numeric column)
if 'Order Date' in df.columns:
    df.sort_values(by='Order Date', inplace=True)
elif 'Total' in df.columns:
    df.sort_values(by='Total', ascending=False, inplace=True)
else:
    df.sort_values(by=df.columns[0], inplace=True)

# 10. Reset the index
df.reset_index(drop=True, inplace=True)

# 11. Save the cleaned and sorted dataset for Power BI
output_path = "/content/cleaned_sorted_pizza_dataset.csv"
df.to_csv(output_path, index=False)

print(f"\n Cleaned & sorted dataset saved to: {output_path}")
```
